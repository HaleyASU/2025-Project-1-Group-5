```python
#Edit the column names to make them easier to read when I concatinate all the DataFrames together
twenty_year_inflation_rate = twenty_year_inflation_rate.rename(columns={"observation_date": "Date", "T20YIEM": "Inflation Rate"})
twenty_year_inflation_rate
```

```python
#We chose not to use this graph because inflation could be better shown on a graph by itself
# Import necessary Dependencies
import matplotlib.pyplot as plt
import pandas as pd

# Convert the 'Date Dispersed' columns to datetime if they aren't already
cleaned_fourty_eight_month_df['Date Dispersed 48mo'] = pd.to_datetime(cleaned_fourty_eight_month_df['Date Dispersed 48mo'])
cleaned_sixty_month_df['Date Dispersed 60mo'] = pd.to_datetime(cleaned_sixty_month_df['Date Dispersed 60mo'])
cleaned_seventy_two_month_df['Date Dispersed 72mo'] = pd.to_datetime(cleaned_seventy_two_month_df['Date Dispersed 72mo'])
twenty_year_inflation_rate['Date'] = pd.to_datetime(twenty_year_inflation_rate['Date'])

# Define a specific date range (last 10 years)
start_date = '2014-01-01'  
end_date = '2025-01-01'   

# Filter the data based on the date range
filtered_fourty_eight_month_df = cleaned_fourty_eight_month_df[(cleaned_fourty_eight_month_df['Date Dispersed 48mo'] >= start_date) & 
                                                                (cleaned_fourty_eight_month_df['Date Dispersed 48mo'] <= end_date)]

filtered_sixty_month_df = cleaned_sixty_month_df[(cleaned_sixty_month_df['Date Dispersed 60mo'] >= start_date) & 
                                                  (cleaned_sixty_month_df['Date Dispersed 60mo'] <= end_date)]

filtered_seventy_two_month_df = cleaned_seventy_two_month_df[(cleaned_seventy_two_month_df['Date Dispersed 72mo'] >= start_date) & 
                                                              (cleaned_seventy_two_month_df['Date Dispersed 72mo'] <= end_date)]

# Merge the inflation rate with the loan dataframes on the date (assuming inflation rate is monthly)
merged_df = pd.merge(filtered_fourty_eight_month_df, twenty_year_inflation_rate, left_on='Date Dispersed 48mo', right_on='Date', how='inner')
merged_df = pd.merge(merged_df, filtered_sixty_month_df, left_on='Date Dispersed 48mo', right_on='Date Dispersed 60mo', how='inner')
merged_df = pd.merge(merged_df, filtered_seventy_two_month_df, left_on='Date Dispersed 48mo', right_on='Date Dispersed 72mo', how='inner')

# Plotting the data
plt.figure(figsize=(12, 8))

# Plot the interest rates
plt.plot(merged_df['Date Dispersed 48mo'], 
         merged_df['Interest Rate 48mo'], 
         label='48-month Loan', color='blue', linestyle='-', marker='o')

plt.plot(merged_df['Date Dispersed 60mo'], 
         merged_df['Interest Rate 60mo'], 
         label='60-month Loan', color='green', linestyle='-', marker='x')

plt.plot(merged_df['Date Dispersed 72mo'], 
         merged_df['Interest Rate 72mo'], 
         label='72-month Loan', color='red', linestyle='-', marker='s')

# Plot inflation rate
plt.plot(merged_df['Date'], merged_df['Inflation Rate'], 
         label='Inflation Rate', color='purple', linestyle='--', marker='^')

# Customize the plot
plt.title('Interest Rates for Different Loan Terms and Inflation Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Rate (%)')
plt.xticks(rotation=45)
plt.legend(title='Legend')
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
