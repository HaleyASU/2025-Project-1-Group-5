```python
# Dependencies
import os 
import csv
import pandas as pd 
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
#Module for reading CSV files
path = os.path.join('Resources', 'average_auto_finance.csv')
path
```

```python
with open(path) as csvfile:
    average_financed = csv.reader(csvfile)  
    print(average_financed)

    # Read the header row first 
    csv_header = next(average_financed)
    print(f"CSV Header: {csv_header}")

    # Read each row of data after the header
    for row in average_financed:
        print(row)
```

```python
#creating a df
average_financed_df = pd.read_csv(path)
average_financed_df.head()
```

```python
#Drop and NaNs
cleaned_average_financed_df = average_financed_df.dropna(how='any')
cleaned_average_financed_df.head()
```

```python
#Edit the column names to make them easier to read when I concatinate all the DataFrames together
cleaned_average_financed_df = cleaned_average_financed_df.rename(columns={"observation_date": "Date Dispersed", "DTCTLVENANM": "Amount Financed"})
cleaned_average_financed_df.head()
```

```python
#Reset the Index
cleaned_average_financed_df = cleaned_average_financed_df.reset_index(drop=True)
cleaned_average_financed_df.head()
```

```python
# Convert the 'Date Dispersed' column to datetime if it's not already
cleaned_average_financed_df['Date Dispersed'] = pd.to_datetime(cleaned_average_financed_df['Date Dispersed'])

# Filter the data to include only dates within the range
start_date = '2015-01-01'
end_date = '2025-01-01'

filtered_df = cleaned_average_financed_df[(cleaned_average_financed_df['Date Dispersed'] >= start_date) & 
                                           (cleaned_average_financed_df['Date Dispersed'] <= end_date)]

# Plotting the filtered data
plt.figure(figsize=(10, 6))
plt.plot(filtered_df['Date Dispersed'], filtered_df['Amount Financed'], label='Amount Financed', marker='o', color='green')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Amount Financed')
plt.title('Average Amount Financed for New Car Loans at Finance Companies')

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Display grid for better readability
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.legend()
plt.savefig('average_financed.png')
plt.show()
```

```python
#creating a df
med_income_df= pd.read_csv(path)
med_income_df.head()
```

```python
#Edit the column names to make them easier to read when I concatinate all the DataFrames together
med_income_df = med_income_df.rename(columns={"observation_date": "Date", "MEHOINUSA672N": "Median Household Income"})
med_income_df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure 'Date' is in datetime format for plotting
med_income_df['Date'] = pd.to_datetime(med_income_df['Date'])

# Define the start and end dates
start_date = pd.to_datetime("2015-01-01")
end_date = pd.to_datetime("2025-01-01")

# Filter the data for the specified date range
filtered_df = med_income_df[(med_income_df['Date'] >= start_date) & (med_income_df['Date'] <= end_date)]

# Plotting the filtered 'Median Household Income' over time
plt.figure(figsize=(10, 6))
plt.plot(filtered_df['Date'], filtered_df['Median Household Income'], marker='o', linestyle='-', color='b')

# Add title and labels
plt.title("Median Household Income Over Time (2015-2025)")
plt.xlabel("Date")
plt.ylabel("Median Household Income")
plt.xticks(rotation=45)

# Ensure labels are not cut off
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# --- Load and clean "Average Auto Finance" data ---
# Load the CSV
path_auto_finance = os.path.join('Resources', 'average_auto_finance.csv')
auto_financed_df = pd.read_csv(path_auto_finance)

# Drop NaN values and clean column names
auto_financed_df = auto_financed_df.dropna(how='any')
auto_financed_df = auto_financed_df.rename(columns={"observation_date": "Date Dispersed", "DTCTLVENANM": "Amount Financed"})

# Convert 'Date Dispersed' to datetime format
auto_financed_df['Date Dispersed'] = pd.to_datetime(auto_financed_df['Date Dispersed'])

# Filter for the date range
start_date = '2015-01-01'
end_date = '2025-01-01'
filtered_auto_financed_df = auto_financed_df[(auto_financed_df['Date Dispersed'] >= start_date) & 
                                              (auto_financed_df['Date Dispersed'] <= end_date)]

# --- Load and clean "Median Household Income" data ---
# Load the CSV
path_household_income = os.path.join('Resources', 'median_household_income.csv')
med_income_df = pd.read_csv(path_household_income)

# Clean column names
med_income_df = med_income_df.rename(columns={"observation_date": "Date", "MEHOINUSA672N": "Median Household Income"})

# Convert 'Date' to datetime format
med_income_df['Date'] = pd.to_datetime(med_income_df['Date'])

# Filter for the date range
filtered_med_income_df = med_income_df[(med_income_df['Date'] >= start_date) & (med_income_df['Date'] <= end_date)]

# --- Plot both datasets on the same graph ---
plt.figure(figsize=(14, 6))

# Plot the "Amount Financed" data (on the left y-axis)
plt.plot(filtered_auto_financed_df['Date Dispersed'], filtered_auto_financed_df['Amount Financed'], 
         marker='o', color='green', label='Amount Financed')

# Plot the "Median Household Income" data (on the same graph, but without a secondary y-axis)
plt.plot(filtered_med_income_df['Date'], filtered_med_income_df['Median Household Income'], 
         marker='o', color='blue', label='Median Household Income')

# Adding title and labels
plt.title("Median Household Income vs. Average Amount Financed (2015-2025)")
plt.xlabel("Date")
plt.ylabel("Amount Financed in Thousands")

# Adding legends
plt.legend(loc='upper left')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display grid for better readability
plt.grid(True)

# Ensure everything fits
plt.tight_layout()
plt.savefig('med_income_to_amount_financed.png')

# Show the plot
plt.show()
```
