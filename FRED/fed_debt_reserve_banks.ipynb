```python
# Dependencies
import os 
import csv
import pandas as pd 
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
#Module for reading CSV files
path = os.path.join('Resources', 'federal_debt_held_by_federal_reserve_banks.csv')
path
```

```python
with open(path) as csvfile:
    fed_reserve_debt= csv.reader(csvfile)  
    print(fed_reserve_debt)

    # Read the header row first 
    csv_header = next(fed_reserve_debt)
    print(f"CSV Header: {csv_header}")

    # Read each row of data after the header
    for row in fed_reserve_debt:
        print(row)
```

```python
#creating a df
fed_reserve_debt_df = pd.read_csv(path)
fed_reserve_debt_df.head()
```

```python
#Edit the column names to make them easier to read when I concatinate all the DataFrames together
fed_reserve_debt_df = fed_reserve_debt_df.rename(columns={"observation_date": "Date", "FDHBFRBN": "Federal Debt in Billions"})
fed_reserve_debt_df.head()
```

```python
# Filter the data to include only dates within the range
start_date = '2015-01-01'
end_date = '2025-01-01'

filtered_df = fed_reserve_debt_df[(fed_reserve_debt_df['Date'] >= start_date) & (fed_reserve_debt_df['Date'] <= end_date)]

# Plotting the filtered data
plt.figure(figsize=(10,6))
plt.plot(filtered_df['Date'], filtered_df['Federal Debt in Billions'], label='Federal Debt', color='blue')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Federal Debt (in Billions)')
plt.title('Federal Debt Held by Federal Reserve Banks (2015-2025)')

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Display grid for better readability
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.legend()
plt.savefig('Federal_Debt.png')
plt.show()
```

```python
#Module for reading CSV files
path = os.path.join('Resources', '20y_inflation_rate.csv')
path
```

```python
with open(path) as csvfile:
    twenty_year_inflation_rate = csv.reader(csvfile)  
    print(twenty_year_inflation_rate)

    # Read the header row first 
    csv_header = next(twenty_year_inflation_rate)
    print(f"CSV Header: {csv_header}")

    # Read each row of data after the header
    for row in twenty_year_inflation_rate:
        print(row)
```

```python
#creating a df
twenty_year_inflation_rate = pd.read_csv(path)
twenty_year_inflation_rate
```

```python
#Edit the column names to make them easier to read when I concatinate all the DataFrames together
twenty_year_inflation_rate = twenty_year_inflation_rate.rename(columns={"observation_date": "Date", "T20YIEM": "Inflation Rate"})
twenty_year_inflation_rate
```

```python

# Convert the 'Date' column to datetime
twenty_year_inflation_rate['Date'] = pd.to_datetime(twenty_year_inflation_rate['Date'])

# Filter the data for the date range 01-01-2015 to 01-01-2025
start_date = pd.to_datetime('2015-01-01')
end_date = pd.to_datetime('2025-01-01')

filtered_data = twenty_year_inflation_rate[(twenty_year_inflation_rate['Date'] >= start_date) & (twenty_year_inflation_rate['Date'] <= end_date)]

# Plot the filtered data
plt.figure(figsize=(10, 6))  # Set figure size
plt.plot(filtered_data['Date'], filtered_data['Inflation Rate'], marker='o', color='b')

# Set title and labels
plt.title('20-Year Inflation Rate Over Time (2015-2025)')
plt.xlabel('Date')
plt.ylabel('Inflation Rate (%)')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()  # Adjust the plot to ensure everything fits
plt.savefig('Inflation_Rate.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Merge the two datasets on 'Date'
merged_df = pd.merge(fed_reserve_debt_df, twenty_year_inflation_rate, on="Date")

# Convert 'Date' column to datetime format
merged_df['Date'] = pd.to_datetime(merged_df['Date'])

# Set the Date column as the index
merged_df.set_index('Date', inplace=True)

# Plot the data using matplotlib
plt.figure(figsize=(10, 6))

# Plot federal debt and inflation rate on the same graph
plt.plot(merged_df.index, merged_df['Federal Debt in Billions'], label="Federal Debt in Billions", color='blue')
plt.plot(merged_df.index, merged_df['Inflation Rate'], label="Inflation Rate", color='red')

# Adding labels and title
plt.title('Correlation Between Federal Debt and Inflation Rate')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
