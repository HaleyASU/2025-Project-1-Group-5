```python
# Dependencies
import os 
import csv
import pandas as pd 
from pathlib import Path
```

```python
#Module for reading CSV files
path = os.path.join('Resources', 'total_vehicle_sales.csv')
path
```

```python
with open(path) as csvfile:
    vehicle_sales = csv.reader(csvfile)  
    print(vehicle_sales)

    # Read the header row first 
    csv_header = next(vehicle_sales)
    print(f"CSV Header: {csv_header}")

    # Read each row of data after the header
    for row in vehicle_sales:
        print(row)
```
